```python
def parse_fasta(filename):
    result = {}
    with open(filename) as f:
        current_id = None
        lines = []
        for l in f:
            l = l.strip()
            if not l:
                continue
            elif l.startswith(">"):
                if current_id:
                    result[current_id] = "".join(lines)
                current_id = l[1:].split("|")[0]
                lines = []
            else:
                lines.append(l)
        if current_id:
            result[current_id] = "".join(lines)
        return result

proteome = parse_fasta("../Figures/COVID/AA_sequence.txt")
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import rc 


def plot_spike_sequence(
        start=None,
        end=None,
        figsize=(20,5),
        letters_per_line=100,
        x_padding_left=.05,
        x_padding_right=.05,
        y_padding_top=.2,
        y_padding_bottom=.2,
        accessibility_threshold=0.3,
        min_accessility_kmer=3,
        fontweight="bold",
        fontsize=12):
    
    S = proteome["S"]
    
    if not start:
        start = 1
    if not end:
        end = len(S)
        
    pre = "%dâ€”" % start
    after = "â€”%d" % end
    S_with_numbers = pre +  S[start - 1:end] + after
    

    
    glycosites = set([])
    with open("../glycosylation/glycosites-Watanabe.txt") as f:
        for l in f:
            l = l.strip()
            if l.startswith("#"):
                continue
            if not l:
                continue
            glycosites.add(int(l))

    antibody_epitope_positions = set([])
    for epitope_start, epitope_end in zip(df_sars2_S.Start, df_sars2_S.End):
        for i in range(epitope_start, epitope_end + 1):
            antibody_epitope_positions.add(i)

    accessible_positions = set([])
    n_accessible = 0
    for aa_num, accessibility in zip(
            df_accessibility["Residue Proper Numbering"],
            df_accessibility["SiteSpecific Accessiblity"]):
        if accessibility > accessibility_threshold:
            n_accessible += 1
        else:
            if n_accessible >= min_accessility_kmer:
                for accessible_aa_num in range(aa_num - n_accessible, aa_num):
                    accessible_positions.add(accessible_aa_num)
            n_accessible = 0
            
    polymorphic_sites = {614}
    
    fig = plt.figure(figsize=figsize)
    plt.xticks([])
    plt.yticks([])


    num_lines = len(S_with_numbers) // letters_per_line + 1

    for i, letter in enumerate(S_with_numbers):
        aa_num = i + start - len(pre)
        line = (i // letters_per_line)
        col = (i % letters_per_line)
        x =  col / letters_per_line

        x_padding_total = x_padding_left + x_padding_right
        x_padded = x * (1.0 - x_padding_total)  + x_padding_left
        y = 1.0 - line  / num_lines
        y_padding_total = y_padding_top + y_padding_bottom
        y_padded = y * (1.0 - y_padding_total) + y_padding_bottom
        
        if i < len(pre):
        
            fontfamily = "serif"
            curr_fontsize = int(0.75 * fontsize)
            color = (0.2, 0.2, 0.2)
            va = "top"
            ha = "center"
        elif i >= (len(S) + len(pre)):
            fontfamily = "serif"
            curr_fontsize = int(0.75 * fontsize)
            color = (0.2, 0.2, 0.2)
            ha = "center"
            va = "top"
        else:
            fontfamily = "monospace"
            curr_fontsize = fontsize
            color = (0, 0, 0, 1)
            ha = "center"
            va = "top"
        
        if aa_num in polymorphic_sites:
            color = (0.9, 0.1, 0.1, 1)
        
        facecolor = (0, 0, 0)
        alpha = 0
        
        if aa_num in glycosites:
            color = (0.1, 0.1, 0.9)
            
        if aa_num in antibody_epitope_positions:
            if aa_num in accessible_positions:
                facecolor = (0.8, 0.8, 0.1)
                alpha = 0.4
            else:
                facecolor = (0.8, 0.8, 0.1)
                alpha = 0.1
        
            
        plt.text(
            x=x_padded, 
            y=y_padded, 
            s=letter, 
            family=fontfamily, 
            size=curr_fontsize, 
            color=color,
            ha=ha, 
            va=va, 
            fontweight=fontweight,
            bbox=dict(boxstyle='square,pad=0.01', facecolor=facecolor, alpha=alpha))
    plt.xlabel(
        "Amino acid sequence of spike glycoprotein with linear antibody epitopes, D614 mutation and glycosites",
        fontsize=16)
```
